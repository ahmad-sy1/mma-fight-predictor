```python
import pandas as pd

# Dataset inladen
df = pd.read_csv('../data/ufc-master.csv')

# Hoeveel rijen en kolommen?
print("Vorm van de dataset:", df.shape)

# Eerste 5 rijen bekijken
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Hoeveel ontbrekende waarden per kolom?
ontbrekend = df.isnull().sum()
ontbrekend[ontbrekend > 0].sort_values(ascending=False)
```

```python
# Volledige lijst met percentages
ontbrekend = df.isnull().sum()
ontbrekend_pct = (ontbrekend / len(df) * 100).round(1)

result = pd.DataFrame({
    'Ontbrekend': ontbrekend,
    'Percentage': ontbrekend_pct
})

result[result['Ontbrekend'] > 0].sort_values('Percentage', ascending=False)
```

```python
# Volledige lijst printen zonder afkapping
pd.set_option('display.max_rows', None)
result[result['Ontbrekend'] > 0].sort_values('Percentage', ascending=False)
```

```python
kolommen_verwijderen = [
    # Ranking kolommen (bijna alles leeg)
    'BWFeatherweightRank', 'RWFeatherweightRank', 'BPFPRank', 'BWFlyweightRank',
    'RWFlyweightRank', 'BWStrawweightRank', 'BWBantamweightRank', 'BWelterweightRank',
    'BLightHeavyweightRank', 'BLightweightRank', 'BBantamweightRank', 'BFeatherweightRank',
    'BFlyweightRank', 'BMiddleweightRank', 'RWStrawweightRank', 'BHeavyweightRank',
    'RWBantamweightRank', 'RFeatherweightRank', 'RLightweightRank', 'RBantamweightRank',
    'RHeavyweightRank', 'RLightHeavyweightRank', 'RMiddleweightRank', 'RWelterweightRank',
    'RFlyweightRank', 'RPFPRank', 'BMatchWCRank', 'RMatchWCRank',
    # Post-fight info
    'Finish', 'FinishDetails', 'FinishRound', 'FinishRoundTime', 'TotalFightTimeSecs',
    # Niet relevant
    'EmptyArena', 'Location', 'Country', 'Date', 'RedFighter', 'BlueFighter',
    # Specifieke odds met veel missende waarden
    'BSubOdds', 'BKOOdds', 'RSubOdds', 'RKOOdds', 'BlueDecOdds', 'RedDecOdds',
    'RedExpectedValue', 'BlueExpectedValue'
]

df_clean = df.drop(columns=kolommen_verwijderen)
print("Kolommen over:", df_clean.shape[1])
print(df_clean.columns.tolist())
```

```python
# Hoeveel ontbrekende waarden zijn er nog?
print("Rijen voor opschonen:", df_clean.shape[0])

# Verwijder rijen met ontbrekende waarden
df_clean = df_clean.dropna()

print("Rijen na opschonen:", df_clean.shape[0])
print("Verwijderd:", 6528 - df_clean.shape[0], "rijen")
```

```python
# Opgeschoonde dataset opslaan
df_clean.to_csv('../data/ufc_clean.csv', index=False)
print("Opgeslagen als ufc_clean.csv")
```
